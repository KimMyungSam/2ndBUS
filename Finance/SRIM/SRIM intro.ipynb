### ROE = 단기순이익 / 자기자본
### = 당기순이익 / 매출액 x 매출액 / 자기자본
### = 당기순이익 / 매출액 x 매출액 / 총자산 x 총자산 / 자기자본
### = 당기순이익 / 매출액 x 매출액 / 총자산 x (1+타인자본/자기자본)
### = 매출액 순이익률 / 총자산회전율 x (1 + 부채비율)

- 업종별로 ROE를 높이는 포인트가 달라지고 기업별 강접도 다르다.

- 사내유보율이 0보다 크고 ROE가 자기자본비용(Ke)보다 높은 경우 기업은 계속 성장하고 있다고 

### RIM = 장부가치 + 초과이익 / (1+Ke) + 잔여가치 / (1+Ke)
#### 초과이익 = 순이익 - 장부가치 x Ke = (ROE-Ke) x 장부가치
#### RIM 기업가치 = 자산가치 + 초과이익의 현재가치
- 초과이익과 잔여가치 계산에서 ROE가 변수이기 때문에 기업의 ROE추세를 파악해 ROE의 지속성 여부를 추정하는 것은 매우 중요하다.
